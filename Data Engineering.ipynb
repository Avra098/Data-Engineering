```python
df.rename(columns={
    'Name': 'Place',
    'Google review rating': 'Rating'
}, inplace=True)
```

```python
required_columns = ['City', 'Place', 'Rating']
if not all(col in df.columns for col in required_columns):
    raise ValueError("Missing required columns in the dataset")
```

```python
def rank_places(city_name):
    df_filtered = df[df['City'].str.lower() != city_name.lower()]
    df_filtered = df_filtered.sort_values(by=['Rating'], ascending=False)
    
    return df_filtered[['Place', 'City', 'Rating']]
```

```python
city_input = "Mumbai"
ranked_places = rank_places(city_input)
print(ranked_places)
```
